```python
#Import some libraries
import PySimpleGUI as sg
from PIL import Image
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import numpy as np
from sympy import S,symbols,printing

#plt.clear()

#Convert to integer
gdpData['index'] = gdpData['index'].astype(int)
population['index'] = population['index'].astype(int)

displayOn = 1

#Default state is Texas
text_input = "Texas"

while (displayOn == 1):
    #Create the unemployment graph
    def create_plot():
        unemployment.plot('index', text_input)
        plt.title('Unemployment Rate Vs Year', fontsize=10)
        plt.xlabel('Year', fontsize=10)
        plt.ylabel('Unemployment Rate', fontsize=10)
        plt.grid(True)
        return plt.gcf()

    #Create the GDP graph
    def create_plot1():
        gdpData.plot('index', text_input)
        plt.title('GDP Vs Year', fontsize=10)
        plt.xlabel('Year', fontsize=10)
        plt.ylabel('GDP', fontsize=10)
        plt.grid(True)
        return plt.gcf()

    #Create the Population graph
    def create_plot2():
        population.plot('index', text_input)
        plt.title('Population Vs Year', fontsize=10)
        plt.xlabel('Year', fontsize=10)
        plt.ylabel('GDP', fontsize=10)
        plt.grid(True)
        return plt.gcf()

    #Create a GDP vs. Unemployment Rate graph
    #https://cmdlinetips.com/2019/10/how-to-make-a-plot-with-two-different-y-axis-in-python-with-matplotlib/
    # create figure and axis objects with subplots()
    fig,ax = plt.subplots()
    # make a plot
    ax.plot(unemployment['index'], unemployment[text_input], color="red", marker="o")
    # set x-axis label
    ax.set_xlabel("year",fontsize=14)
    # set y-axis label
    ax.set_ylabel("Unemployment Rate",color="red",fontsize=14)
    # twin object for two different y-axis on the sample plot
    ax2=ax.twinx()
    # make a plot with different y-axis using second axis object
    ax2.plot(unemployment['index'], gdpData[text_input],color="blue",marker="o")
    ax2.set_ylabel("GDP",color="blue",fontsize=14)
    plt.show()
    # save the plot as a file
    fig.savefig('image.png',
                format='png',
                dpi=100,
                bbox_inches='tight')


    #Projecting GDP
    gdp1 = gdpData[text_input].values[0]
    gdp2 = gdpData[text_input].values[6]
    gdpSlope = (gdp2-gdp1)/(2019-2013)
    gdpSlope = round(gdpSlope, 3)

    #Projecting Unemployment Rate
    unemployment1 = unemployment[text_input].values[0]
    unemployment2 = unemployment[text_input].values[6]
    unemploymentSlope = (unemployment2-unemployment1)/(2019-2013)
    unemploymentSlope = round(unemploymentSlope, 3)

    #Predictions text
    predictions = "The GDP is expected to change by " + str(gdpSlope) + " million dollars each year. The unemployment rate is expected to change by " + str(unemploymentSlope) + " percent each year"

    #Calculating mean GDP
    meanGDP = str(round(gdpData[text_input].mean(),3))

    #Calculating mean UER
    meanUER = str(round(unemployment[text_input].mean(),3))

    #Mean text
    meanText = "The mean GDP is " + meanGDP + " million dollars. The mean Unemployment Rate is " + meanUER

    #https://stackoverflow.com/questions/67079155/displaying-an-image-using-pysimplegui-without-having-to-use-an-event-listener
    #sg.Image('image.jpeg')

    #Create the simple GUI layout
    layout = [[sg.Text(text_input + ' State Data')],
                    [sg.Canvas(size=(1350, 400), key='-CANVAS-')],
                    [sg.Text(predictions)],
                    [sg.Text(meanText)],
                    [sg.InputText()],
                    [sg.Button('Submit'),sg.Exit()],
                    [sg.Image(size=(1000,350),filename="image.png")]]

    #Create the standard line graph function
    def draw_figure(canvas, figure):
        figure_canvas_agg = FigureCanvasTkAgg(figure, canvas)
        figure_canvas_agg.draw()
        figure_canvas_agg.get_tk_widget().pack(side='left', fill='both', expand=1)
        return figure_canvas_agg

    #Create the window
    window = sg.Window('Aarian Dhanani Data Science Final Project', layout, finalize=True, element_justification='center')

    #Create the graphs
    draw_figure(window['-CANVAS-'].TKCanvas, create_plot())
    draw_figure(window['-CANVAS-'].TKCanvas, create_plot1())
    draw_figure(window['-CANVAS-'].TKCanvas, create_plot2())

    #Refresh if a button is clicked
    event, values = window.read()    
    window.close()

    text_input = values[0]

    #Close if Exit is clicked
    while True:
        event, values = window.read()
        if event == sg.WIN_CLOSED or event == 'Exit':
            break

    window.close()
```
