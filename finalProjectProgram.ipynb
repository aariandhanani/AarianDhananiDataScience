```python
import PySimpleGUI as sg
from PIL import Image
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg

#plt.clear()

gdpData['index'] = gdpData['index'].astype(int)
population['index'] = population['index'].astype(int)

displayOn = 1
text_input = "Texas"
while (displayOn == 1):

    def create_plot():
        unemployment.plot('index', text_input)
        plt.title('Unemployment Rate Vs Year', fontsize=10)
        plt.xlabel('Year', fontsize=10)
        plt.ylabel('Unemployment Rate', fontsize=10)
        plt.grid(True)
        return plt.gcf()

    def create_plot1():
        gdpData.plot('index', text_input)
        plt.title('GDP Vs Year', fontsize=10)
        plt.xlabel('Year', fontsize=10)
        plt.ylabel('GDP', fontsize=10)
        plt.grid(True)
        return plt.gcf()

    def create_plot2():
        population.plot('index', text_input)
        plt.title('Population Vs Year', fontsize=10)
        plt.xlabel('Year', fontsize=10)
        plt.ylabel('GDP', fontsize=10)
        plt.grid(True)
        return plt.gcf()

    #https://cmdlinetips.com/2019/10/how-to-make-a-plot-with-two-different-y-axis-in-python-with-matplotlib/
    # create figure and axis objects with subplots()
    fig,ax = plt.subplots()
    # make a plot
    ax.plot(unemployment['index'], unemployment[text_input], color="red", marker="o")
    # set x-axis label
    ax.set_xlabel("year",fontsize=14)
    # set y-axis label
    ax.set_ylabel("GDP",color="red",fontsize=14)

    # twin object for two different y-axis on the sample plot
    ax2=ax.twinx()
    # make a plot with different y-axis using second axis object
    ax2.plot(unemployment['index'], gdpData[text_input],color="blue",marker="o")
    ax2.set_ylabel("Unemployment Rate",color="blue",fontsize=14)
    plt.show()
    # save the plot as a file
    fig.savefig('image.png',
                format='png',
                dpi=100,
                bbox_inches='tight')

    #https://stackoverflow.com/questions/67079155/displaying-an-image-using-pysimplegui-without-having-to-use-an-event-listener
    #sg.Image('image.jpeg')

    layout = [[sg.Text(text_input + ' State Data')],
                    [sg.Canvas(size=(1500, 1000), key='-CANVAS-')], [sg.InputText()], [sg.Button('Submit')],
                    [sg.Exit()], [sg.Image(size=(1000,350),filename="image.png")]]

    def draw_figure(canvas, figure):
        figure_canvas_agg = FigureCanvasTkAgg(figure, canvas)
        figure_canvas_agg.draw()
        figure_canvas_agg.get_tk_widget().pack(side='left', fill='both', expand=1)
        return figure_canvas_agg

    window = sg.Window('Demo Application - Embedding Matplotlib In PySimpleGUI', layout, finalize=True, element_justification='center')

    draw_figure(window['-CANVAS-'].TKCanvas, create_plot())
    draw_figure(window['-CANVAS-'].TKCanvas, create_plot1())
    draw_figure(window['-CANVAS-'].TKCanvas, create_plot2())

    #
    event, values = window.read()    
    window.close()

    text_input = values[0]
    #

    while True:
        event, values = window.read()
        if event == sg.WIN_CLOSED or event == 'Exit':
            break

    window.close()
```
